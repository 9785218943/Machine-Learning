## Q1- What is data encoding? How is it useful in data science?

Ans- Encoding is the process of converting the data or a given sequence of characters,symbols,alphabets etc.into a specified format,for the secured transmission of data.Decoding is the reverse process of encoding which is to extract the information from the converted format.

data preparation is a mandatory task. There are various tasks we require to perform in the data preparation. Encoding categorical data is one of such tasks which is considered crucial. As we know, most of the data in real life come with categorical string values and most of the machine learning models work with integer values only and some with other different values which can be understandable for the model. All models basically perform mathematical operations which can be performed using different tools and techniques. But the harsh truth is that mathematics is totally dependent on numbers. So in short we can say most of the models require numbers as the data, not strings or not anything else and these numbers can be float or integer. 

## Q2-What is nominal encoding? Provide an example of how you would use it in a real-world scenario.

Ans-Nominal encoding is used to transform categorical data into numerical data by assigning a unique numerical value to each category in a feature.

Nominal data: This type of categorical data consists of the name variable without any numerical values. For example, in any organization, the name of the different departments like research and development department, human resource department, accounts and billing department etc.



```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
## create  a simple dataframe
df=pd.DataFrame({'color':['red','blue','green','purple','red','green','blue']})
```

```python
df.head()
```

```python
## create an instance of Onehotencoder
encoder=OneHotEncoder()
```

```python
## perform fit and transform
encoded=encoder.fit_transform(df[['color']]).toarray()
```

```python
import pandas as pd
encoder_df=pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
```

```python
encoder_df
```

```python
import pandas as pd
import numpy as np

# Define the headers since the data does not have any
headers = ["symboling", "normalized_losses", "make", "fuel_type", "aspiration",
           "num_doors", "body_style", "drive_wheels", "engine_location",
           "wheel_base", "length", "width", "height", "curb_weight",
           "engine_type", "num_cylinders", "engine_size", "fuel_system",
           "bore", "stroke", "compression_ratio", "horsepower", "peak_rpm",
           "city_mpg", "highway_mpg", "price"]

# Read in the CSV file and convert "?" to NaN
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data",
                  header=None, names=headers, na_values="?" )
df.head()
```

```python
df.dtypes
```

```python
obj_df = df.select_dtypes(include=['object']).copy()
obj_df.head()
```

```python
obj_df[obj_df.isnull().any(axis=1)]
```

```python
obj_df["num_doors"].value_counts()
```

```python
obj_df = obj_df.fillna({"num_doors": "four"})
```

```python
obj_df["body_style"] = obj_df["body_style"].astype('category')
obj_df.dtypes
```

```python
obj_df["body_style_cat"] = obj_df["body_style"].cat.codes
obj_df.head()
```

```python
pd.get_dummies(obj_df, columns=["drive_wheels"]).head()
```

```python
pd.get_dummies(obj_df, columns=["body_style", "drive_wheels"], prefix=["body", "drive"]).head()
```

```python
from sklearn.preprocessing import OrdinalEncoder

ord_enc = OrdinalEncoder()
obj_df["make_code"] = ord_enc.fit_transform(obj_df[["make"]])
obj_df[["make", "make_code"]].head(11)
```

```python
from sklearn.preprocessing import OneHotEncoder

oe_style = OneHotEncoder()
oe_results = oe_style.fit_transform(obj_df[["body_style"]])
pd.DataFrame(oe_results.toarray(), columns=oe_style.categories_).head()
```

```python
obj_df = obj_df.join(pd.DataFrame(oe_results.toarray(), columns=oe_style.categories_))
```
