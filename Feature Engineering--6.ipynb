## Q2. Spearman's rank correlation is a measure of the monotonic relationship between two variables.
# Suppose you have collected data on the amount of sleep individuals get each night and their overall job
# satisfaction level on a scale of 1 to 10. Calculate the Spearman's rank correlation between these two
# variables and interpret the result.

Ans-Spearman's rank correlation measures the strength and direction of association between two ranked variables. It basically gives the measure of monotonicity of the relation between two variables i.e. how well the relationship between two variables could be represented using a monotonic function

```python
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
df.head()
```

```python
df.corr()
```

```python
## spearman rank correlation
df.corr(method='spearman')
```
